```python
import pandas as pd

data = pd.DataFrame({
    'view_type': ['external'],
    'angle': ['front_right'],
    'part': ['Tire'],
    'damage': ['glass shatter'],  
    'airbag': ['not_applicable']
})
```

```python
import xgboost as xgb
loaded_model = xgb.XGBClassifier(
    n_estimators=200,
    learning_rate=0.1,
    max_depth=6,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42
)
loaded_model.load_model("xgb_model.json")
```

```python
import pickle
with open('feature_encoders.pkl', 'rb') as f:
    feature_encoders = pickle.load(f)
for col in ['view_type', 'angle', 'part', 'damage', 'airbag']:
    if col in feature_encoders:
        le = feature_encoders[col]
        data[col] = le.transform(data[col])
```

```python
with open('severity_encoder.pkl', 'rb') as f:
    severity_encoder = pickle.load(f)
y_pred_encoded = loaded_model.predict(data)
y_pred_decoded = severity_encoder.inverse_transform(y_pred_encoded)
print("Predicted Severity:", y_pred_decoded[0])
```
