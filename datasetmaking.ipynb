```python
import pandas as pd
import random

view_types = ["internal", "external"]
angles = ["front", "rear", "left", "right", "front_left", "front_right", "rear_left", "rear_right"]

vehicle_parts = [
    "Battery Tester",
    "Cabin Air Filter",
    "Gauge Brakes",
    "Gauge Dial",
    "Gauge Digital",
    "Gauge Manual",
    "Headlight",
    "License Plate",
    "Screen",
    "Tire",
    "Wiper"
]

damage_types = ["none", "scratch", "dent", "crack", "glass shatter", "lamp broken", "tire flat"]
airbag_status = ["deployed", "not_applicable"]

severity_levels = ["undamaged", "minor", "moderate", "severe"]


def estimate_severity(view_type, part, damage, airbag):
    if damage == "none":
        return "undamaged"
    if airbag == "deployed":
        return "severe"
    if damage in ["glass-shatter", "crack"]:
        return "severe"
    if damage in ["dent", "lamp-broken", "tire-flat"]:
        return "moderate"
    if damage == "scratch":
        return "minor"
    return "moderate"


data = []
for i in range(1000):  # generate 1000 rows
    view = random.choice(view_types)
    part = random.choice(vehicle_parts)
    angle = random.choice(angles)
    damage = random.choice(damage_types)
    airbag = "deployed" if (view == "internal" and random.random() < 0.2) else "not_applicable"
    severity = estimate_severity(view, part, damage, airbag)
    
    data.append({
        
        "view_type": view,
        "angle": angle,
        "part": part,
        "damage": damage,
        "airbag": airbag,
        "severity": severity
    })

df = pd.DataFrame(data)
df.to_csv("vehicle_damage_dataset.csv", index=False)
print(" Dataset generated and saved as vehicle_damage_dataset.csv")
df
```
